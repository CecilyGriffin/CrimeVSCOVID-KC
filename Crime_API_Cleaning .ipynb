```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import pandas as pd
from sodapy import Socrata
from sqlalchemy import create_engine
```

```python
client = Socrata("data.kcmo.org", None)
```

```python
# API Calls
crimeresults = client.get("vsgj-uufz", limit=65000)
```

```python
results_df = pd.DataFrame.from_records(crimeresults)
results_df
```

```python
results_df.dropna(how='all')
```

```python
results_df.columns
```

```python
importantcolumns = ['firearm_used_flag','zip_code','report_no']
```

```python
for column in importantcolumns:
    print(column)
    print(results_df[column].value_counts())
    print(f"\n")
```

```python
df = results_df[importantcolumns]
firearmtrue = df.loc[df['firearm_used_flag'] == True]
firearmtrue
```

```python
for column in importantcolumns:
    print(column)
    print(firearmtrue[column].value_counts())
    print(f"\n")
```

```python
firearmtrue.dropna(how = 'all')
```

```python
firearmtrue.to_csv ('NewCrime.csv')
```
