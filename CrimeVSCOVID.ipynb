```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import crime data CSV
crime_file = "Resources/KCPD_Crime_Data.csv"
crime_df = pd.read_csv(crime_file)
crime_df.head()
```

```python
#import COVID data CSV
covid_file = "Resources/COVID-Data.csv"
covid_df = pd.read_csv(covid_file)
covid_df
```

```python
crime_df.reset_index(level=None, drop=False, inplace=True, col_level=0, col_fill='')
```

```python
crime_cols = ['index',"Offense", "Zip Code", "Sex", "Age"]
crime_transformed= crime_df[crime_cols].copy()

# Rename the column headers
crime_transformed = crime_transformed.rename(columns={'index':'id',"Offense": "offense",
                                                          "Zip Code": "zip_code",
                                                          "Sex": "gender","Age":"age"})
```

```python
#drop blank zip codes (99999)
crime_new = crime_transformed.loc[crime_transformed['zip_code'] != '99999']
crime_new
```

```python
#drop NaN
crime_new=crime_new.dropna(how='any')
crime_new
```

```python
#establish database conntection
connection_string = "postgres:Tc1234@localhost:5432/ETL_DB"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
crime_new.to_sql(name='crime', con=engine, if_exists='append', index=False)
```

```python
covid_df.reset_index(level=None, drop=False, inplace=True, col_level=0, col_fill='')
```

```python
covid_cols = ["index","ZipCode", "COVID19 Cases", "Crude Rate per 100,000"]
covid_transformed= covid_df[covid_cols].copy()

# Rename the column headers
covid_transformed = covid_transformed.rename(columns={"index":"id","ZipCode": "zip_code",
                                                          "COVID19 Cases": "covid19_cases",
                                                          "Crude Rate per 100,000": "crude_rate_per_hundred_thousand"})
```

```python
covid_transformed.to_sql(name='covid', con=engine, if_exists='append', index=False)
```
